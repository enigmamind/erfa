```python
import pandas as pd
import json
import re

def extract_weth_usdt_and_ratio(token_transferred):
    if not isinstance(token_transferred, str):
        return pd.Series({'WETH_Amount': 0, 'USDT_Amount': 0, 'Ratio': 0})

    try:
        data = json.loads(token_transferred)[0]['Token_Transferred']
        weth_match = re.search(r'For([\d.]+)\(\$[\d.]+\)Wrapped Ethe... \(WETH\)', data)
        usdt_match = re.search(r'For([\d.]+)\(\$[\d.]+\)Tether USD \(USDT\)', data)

        weth_amount = float(weth_match.group(1)) if weth_match else 0
        usdt_amount = float(usdt_match.group(1)) if usdt_match else 0

        # Calculate ratio (USDT per WETH)
        ratio = abs(usdt_amount / weth_amount) if weth_amount != 0 else 0

        return pd.Series({'WETH_Amount': weth_amount, 'USDT_Amount': usdt_amount, 'Ratio': ratio})
    except (json.JSONDecodeError, KeyError, IndexError, AttributeError):
        return pd.Series({'WETH_Amount': 0, 'USDT_Amount': 0, 'Ratio': 0})

# Load your data
df = pd.read_csv('test-data.csv', parse_dates=['Date'])

# Apply the extraction and ratio calculation function
df[['WETH_Amount', 'USDT_Amount', 'Ratio']] = df['Token_Transferred'].apply(extract_weth_usdt_and_ratio)

# Display the result
print(df[['Transanction_Hash', 'Date', 'WETH_Amount', 'USDT_Amount', 'Ratio']])

# Save the updated dataframe to a new CSV file
df.to_csv('updated_swap_data_with_ratio.csv', index=False)
```

```python
import pandas as pd
import json
import re
from datetime import datetime, timedelta

def extract_weth_usdt_and_ratio(token_transferred):
    if isinstance(token_transferred, str):
        try:
            data = json.loads(token_transferred)[0]['Token_Transferred']
            weth_match = re.search(r'For([\d.]+)\(\$[\d.]+\)Wrapped Ethe... \(WETH\)', data)
            usdt_match = re.search(r'For([\d.]+)\(\$[\d.]+\)Tether USD \(USDT\)', data)

            weth_amount = float(weth_match.group(1)) if weth_match else 0
            usdt_amount = float(usdt_match.group(1)) if usdt_match else 0

            ratio = abs(usdt_amount / weth_amount) if weth_amount != 0 else 0

            return pd.Series({'WETH_Amount': weth_amount, 'USDT_Amount': usdt_amount, 'Ratio': ratio})
        except (json.JSONDecodeError, IndexError):
            return pd.Series({'WETH_Amount': 0, 'USDT_Amount': 0, 'Ratio': 0})
    else:
        return pd.Series({'WETH_Amount': 0, 'USDT_Amount': 0, 'Ratio': 0})

def round_to_5min(dt):
    if pd.isnull(dt):
        return pd.NaT
    return dt - timedelta(minutes=dt.minute % 5, seconds=dt.second, microseconds=dt.microsecond)

# Load your data
df = pd.read_csv('test-data.csv', parse_dates=['Date'])

# Apply the extraction and ratio calculation function
df[['WETH_Amount', 'USDT_Amount', 'Ratio']] = df['Token_Transferred'].apply(extract_weth_usdt_and_ratio)

# Remove rows where WETH_Amount, USDT_Amount, and Ratio are all zero
df = df[(df['WETH_Amount'] != 0) | (df['USDT_Amount'] != 0) | (df['Ratio'] != 0)]

# Create 5-minute bins
df['Five_Minute_Bin'] = df['Date'].apply(round_to_5min)

# Remove rows with NaT values in Five_Minute_Bin
df = df.dropna(subset=['Five_Minute_Bin'])

# Group by 5-minute bins and calculate aggregates
grouped = df.groupby('Five_Minute_Bin').agg({
    'WETH_Amount': 'sum',
    'USDT_Amount': 'sum',
    'Ratio': 'mean',
    'Transanction_Hash': 'count'
}).reset_index()

# Rename columns for clarity
grouped.columns = ['Five_Minute_Bin', 'Total_WETH', 'Total_USDT', 'Average_Ratio', 'Number_Tx']

# Display the result
print(grouped)

# Save the aggregated data to a new CSV file
grouped.to_csv('aggregated_swap_data_5min.csv', index=False)
```

CALCULATE AVG RATIO AND OVERALL RATIO

```python
import pandas as pd
import json
import re
from datetime import datetime, timedelta

def extract_weth_usdt_and_ratio(token_transferred):
    if isinstance(token_transferred, str):
        try:
            data = json.loads(token_transferred)[0]['Token_Transferred']
            weth_match = re.search(r'For([\d.]+)\(\$[\d.]+\)Wrapped Ethe... \(WETH\)', data)
            usdt_match = re.search(r'For([\d.]+)\(\$[\d.]+\)Tether USD \(USDT\)', data)

            weth_amount = float(weth_match.group(1)) if weth_match else 0
            usdt_amount = float(usdt_match.group(1)) if usdt_match else 0

            ratio = usdt_amount / weth_amount if weth_amount != 0 else 0

            return pd.Series({'WETH_Amount': weth_amount, 'USDT_Amount': usdt_amount, 'Ratio': ratio})
        except (json.JSONDecodeError, IndexError):
            return pd.Series({'WETH_Amount': 0, 'USDT_Amount': 0, 'Ratio': 0})
    else:
        return pd.Series({'WETH_Amount': 0, 'USDT_Amount': 0, 'Ratio': 0})

def round_to_5min(dt):
    if pd.isnull(dt):
        return pd.NaT
    return dt - timedelta(minutes=dt.minute % 5, seconds=dt.second, microseconds=dt.microsecond)

# Load your data
df = pd.read_csv('UNISWAP-DATA.csv', parse_dates=['Date'])

# Apply the extraction and ratio calculation function
df[['WETH_Amount', 'USDT_Amount', 'Ratio']] = df['Token_Transferred'].apply(extract_weth_usdt_and_ratio)

# Create 5-minute bins
df['Five_Minute_Bin'] = df['Date'].apply(round_to_5min)

# Remove rows with NaT values in Five_Minute_Bin
df = df.dropna(subset=['Five_Minute_Bin'])

# Group by 5-minute bins and calculate aggregates
grouped = df.groupby('Five_Minute_Bin').agg({
    'WETH_Amount': 'sum',
    'USDT_Amount': 'sum',
    'Ratio': lambda x: sum(x[x != 0]) / len(x[x != 0]) if len(x[x != 0]) > 0 else 0,  # Average of non-zero ratios
    'Transanction_Hash': 'count'
}).reset_index()

# Rename columns for clarity
grouped.columns = ['Five_Minute_Bin', 'Total_WETH', 'Total_USDT', 'Average_Ratio', 'Number_Tx']

# Calculate the overall ratio for each bin
grouped['Overall_Ratio'] = grouped['Total_USDT'] / grouped['Total_WETH']

# Display the result
print(grouped)

# Save the aggregated data to a new CSV file
grouped.to_csv('aggregated_swap_data_5min.csv', index=False)
```
